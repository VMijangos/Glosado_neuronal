# Ploteo de embeddings de modelos

```python
from pickle import load
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from operator import itemgetter
```

```python
#Funcion para plotear los datos con labels
def plot_words(Z,ids):
    #Reduce la dimensionalidad a 2
    Z = PCA(2).fit_transform(Z)
    
    #Plotea con la marcas (marker) y el color indicado (c)
    r=0
    plt.scatter(Z[:,0],Z[:,1], marker='o', c='blue', s=0)
    for label,x,y in zip(ids, Z[:,0], Z[:,1]):
        plt.annotate(label, xy=(x,y), xytext=(-1,1), textcoords='offset points', ha='center', va='bottom')
        r+=1
```

```python
embs_biLSTM = load( open( "Embs.biLSTM.justChar.p", "rb" ) )
embs_biRNN = load( open( "Embs.biRNN.justChar.p", "rb" ) )
```

```python
labels_biLSTM = list(embs_biLSTM.keys())[:-1]
vecs_biLSTM = np.array([np.array(v) for v in embs_biLSTM.values()])[:-1]

labels_biRNN = list(embs_biRNN.keys())[:-1]
vecs_biRNN = np.array([np.array(v) for v in embs_biRNN.values()])[:-1]
```

```python
plot_words(vecs_biLSTM, labels_biLSTM)
plt.title('Embeddings from the biLSTM model')
plt.show()
```

```python
plot_words(vecs_biRNN, labels_biRNN)
plt.title('Embeddings from the bidirectional RNN model')
plt.show()
```
